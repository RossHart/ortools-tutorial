```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.rcParams.update({'font.size': 16})
```

### Build up the plot over time

```python
def setup_axes():
    fig, ax = plt.subplots(1, 1, figsize=(10, 10))
    plt.xlim(0, 30)
    plt.ylim(0, 30)
    plt.xlabel("$y_s$")
    plt.ylabel("$y_p$")
    fig.tight_layout()
    return fig, ax
```

```python
fig, ax = setup_axes()

plt.sca(ax)

# constraint 1
plt.axvline(10, color="#00539f", lw=2, label="constraints")

# constraint 2
plt.axhline(0, color="#00539f", lw=2)

# consraint 3
x_c = [0, 30]
y_c = [(40 - 1.5 * x) / 2 for x in x_c]
plt.plot(x_c, y_c, color="#00539f", lw=2)

# show shaded region
x_f = [10, 30]
y_f = [(40 - 1.5 * x) / 2 for x in x_f]
plt.fill_between(x_f, 0, y_f, color="#00539f", alpha=.2, label="feasible region")
plt.legend()
plt.savefig("./LP_no_obj.png")

# show lines of constant objective
for i, P in enumerate(range(1000, 4001, 1000)):
    x_o = [0, 30]
    y_o = [(P - 125 * x) / 200 for x in x_o]
    
    if i == 0:
        label="constant objective"
    else:
        label = None
    plt.plot(x_o, y_o, color="#ee1c2e", linestyle="dashed", lw=1.5, label=label)

plt.legend()
plt.savefig("./LP.png")

for x in range(10, 30):
    for y in range(30):
        if 1.5 * x + 2.0 * y <= 40:
            plt.scatter(x, y, color="k", marker="o", s=15)

plt.scatter(None, None, color="k", marker="o", s=15, label="allowed values")
            
plt.legend()
plt.savefig("./MILP.png")
```
